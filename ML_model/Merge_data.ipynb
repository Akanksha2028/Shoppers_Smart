```python
import pandas as pd
```

```python
customers = pd.read_csv("C:/Users/akank/intership-june2025/ShoppersSmart/data/customers.csv")
products = pd.read_csv("C:/Users/akank/intership-june2025/ShoppersSmart/data/Products.csv")
transactions = pd.read_csv("C:/Users/akank/intership-june2025/ShoppersSmart/data/Transactions.csv")
```

```python
print("Customers:")
display(customers.head())

print("Products:")
display(products.head())

print("Transactions:")
display(transactions.head())
```

```python
transactions.columns = transactions.columns.str.strip()
products.columns = products.columns.str.strip()
customers.columns = customers.columns.str.strip()
```

```python
print(products.columns)
```

```python
merged_df = pd.merge(transactions, products[['product_id', 'product_name']], on='product_id', how='left')
```

```python
final_df = pd.merge(merged_df, customers[['customer_id', 'email', 'phone_number']], on='customer_id', how='left')
```

```python
trans_cust_df = pd.merge(transactions, customers, on="customer_id", how="left")
```

```python
final_df = pd.merge(trans_cust_df, products, on="product_id", how="left")
```

```python
final_df = final_df[[
    "transaction_id", "customer_id", "product_id", "product_name",
    "purchase_date", "location", "price", "email", "phone_number"
]]
```
