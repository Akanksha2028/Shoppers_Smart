```python
import pandas as pd
```

```python
customers = pd.read_csv("C:/Users/akank/intership-june2025/ShoppersSmart/data/customers.csv")
products = pd.read_csv("C:/Users/akank/intership-june2025/ShoppersSmart/data/Products.csv")
transactions = pd.read_csv("C:/Users/akank/intership-june2025/ShoppersSmart/data/Transactions.csv")
```

```python
print("Customers:")
display(customers.head())

print("Products:")
display(products.head())

print("Transactions:")
display(transactions.head())
```

```python
merged = transactions.merge(customers, on='customer_id', how='left')
merged = merged.merge(products, on='product_id', how='left')
```

```python
print("Merged Dataset Preview:")
display(merged.head())
```

```python
merged.to_csv("C:/Users/akank/intership-june2025/ShoppersSmart/merged_dataset.csv", index=False)
print(" Merged dataset saved as 'merged_dataset.csv'")
```
